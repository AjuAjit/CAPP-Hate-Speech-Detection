### STORING NECESSARY DETAILS IN dataframe

```python
ids = []
retweets = []
hashtags_list = []
tweet_content = []
user_ids =[]
names = []


for tweet in timeline:
    
    user_ids.append(tweet._json['user']['id'])
    names.append(tweet._json['user']['name'])
    ids.append(tweet._json['id'])
    
    hashtags = []
    temp_hash = tweet._json['entities']['hashtags']
    for j in range(0, len(temp_hash)):
        hashtags.append(temp_hash[j]['text'])     
    hashtags_list.append(hashtags)
    
    if 'retweeted_status' in tweet._json:
        tweet_content.append(tweet._json['retweeted_status']['full_text'])
    else:
        tweet_content.append(tweet._json['full_text'])
        
data = {'user_id' : user_ids, 'names' : names, 'tweet_id':ids, 'tweet': tweet_content, 'hashtags' : hashtags_list} 
df = pd.DataFrame(data)
df
```
