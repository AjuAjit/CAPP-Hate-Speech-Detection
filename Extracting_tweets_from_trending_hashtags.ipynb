### Extracting tweets and storing them in a dataframe with user id and tweet id

```python
ids = []
likes = []
retweets = []
hashtags_list = []
tweet_content = []
user_ids =[]
names = []

for i in range(0, len(search_results['statuses'])):
    id = search_results['statuses'][i]['id']
    ids.append(id)
    #tweet = auth_api.get_status(id)
    print(i)
    #likes.append(tweet.favorite_count)
    #retweets.append(tweet.retweet_count)
    user_ids.append(search_results['statuses'][i]['user']['id'])
    names.append(search_results['statuses'][i]['user']['name'])

    
    hashtags = []
    temp_hash = search_results['statuses'][i]['entities']['hashtags']
    for j in range(0, len(temp_hash)):
        hashtags.append(temp_hash[j]['text'])     
    hashtags_list.append(hashtags)
    
    if 'retweeted_status' in search_results['statuses'][i]:
        tweet_content.append(search_results['statuses'][i]['retweeted_status']['full_text'])
    else:
        tweet_content.append(search_results['statuses'][i]['full_text'])
    #print(i)
#print(len(likes))
#print(len(retweets))
#print(hashtags_list)

data = {'user_id' : user_ids, 'names' : names, 'tweet_id':ids, 'tweet': tweet_content, 'hashtags' : hashtags_list} 
df = pd.DataFrame(data)
df
```
